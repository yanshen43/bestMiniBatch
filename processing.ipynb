```python
import cPickle
with open('distance_m2.pkl', 'rb') as f:
    data = cPickle.load(f)
    print data
```

```python
import csv 

with open('distance_m2.csv','wb') as csvFile:
    writer = csv.writer(csvFile, delimiter=',')
    for line in data:
        writer.writerow(line)
```
